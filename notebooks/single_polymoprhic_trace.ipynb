```julia
using LinearAlgebra, Jevo, Jedi, Plots, LambertW, Distributions
Jedi.default_gr!()

# Parameters
generations = 2 * 10^8
ϵ = 2
l_0 = 10
N = 100
nu = 1/N
emat = ϵ * (ones(4, 4) - Matrix{Float64}(I, 4, 4))
n = 4
f0 = 600/2N
rho = 10
```

```julia
l0_kappa(kappa, l) = 1/2 * lambertw(2 * ϵ^2 * N * l * f0 * (n-1)/n^2 * exp(10)/(1+kappa))
fl(l_opt) = l0_kappa(0, 10)/10^2 * n^2 / (n-1) * 1/ϵ
fl_new(l_opt) = l0_kappa(0, 10)^2 / l_opt^2 * n^2 / (n-1) * 1/2
```

```julia
f = Jevo.fermi_fitness(f0=f0, fl=fl_new(10)/2N, epsilon=ϵ)
```

```julia
pop = Jevo.driver_trailer_l(N=N, L=50, l_0=l_0)
Jevo.initiate!(pop, emat)

l_arr = zeros(generations)
for j in 1:generations
    if mean(pop.l) < 6
        pop = Jevo.driver_trailer_l(N=N, L=50, l_0=l_0)
        Jevo.initiate!(pop, emat)
    end
    for m in 1:rand(Poisson(1/10), 1)[1]
        Jevo.mutation!(pop)
    end
    for m in 1:rand(Poisson(1/N^2), 1)[1]
        Jevo.length_mutation!(pop)
    end
    for m in 1:rand(Poisson(rho / 10N), 1)[1]
        Jevo.driver_mutation!(pop)
    end
    Jevo.sample_gen!(pop, f, emat; remove=true)
    l_arr[j] = mean(pop.l)
end
```

```julia
plot(l_arr[1:1000:end])
```
