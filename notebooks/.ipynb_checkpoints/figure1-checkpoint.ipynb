```julia
using Jedi, LinearAlgebra
```

```julia
emat = 2 * (ones(4, 4) - Matrix{Float64}(I, 4, 4))
```

```julia
Est(l) = 2*(3/4 * l - 5)

N = 1000
f0 = 20/2N
f = fermi_fitness(f0=f0, E_Star=Est)
```

```julia
pop = driver_trailer_l(N=10000, L=30, l_0=10)
initiate_rand!(pop, 1)
for i in 1:10000
    Jedi.bp_substitution!(pop, emat, f)
end
```
